```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn import preprocessing

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score

```

```python
sd=pd.read_csv("SalaryData_Train(1).csv")
sd
```

```python
label_encoder=preprocessing.LabelEncoder()
sd["workclass"]=label_encoder.fit_transform(sd["workclass"])
sd["education"]=label_encoder.fit_transform(sd["education"])
sd["maritalstatus"]=label_encoder.fit_transform(sd["maritalstatus"])
sd["occupation"]=label_encoder.fit_transform(sd["occupation"])
sd["relationship"]=label_encoder.fit_transform(sd["relationship"])
sd["race"]=label_encoder.fit_transform(sd["race"])
sd["sex"]=label_encoder.fit_transform(sd["sex"])

sd["native"]=label_encoder.fit_transform(sd["native"])

sd["Salary"]=label_encoder.fit_transform(sd["Salary"])
sd
```

```python
sdtes=pd.read_csv("SalaryData_Test(1).csv")
sdtes
```

```python
x_train=sd.iloc[0:1000,0:13]
x_train
```

```python
y_train=sd.iloc[0:1000,13]
y_train
```

```python
sdtes["workclass"]=label_encoder.fit_transform(sdtes["workclass"])
sdtes["education"]=label_encoder.fit_transform(sdtes["education"])
sdtes["maritalstatus"]=label_encoder.fit_transform(sdtes["maritalstatus"])
sdtes["occupation"]=label_encoder.fit_transform(sdtes["occupation"])
sdtes["relationship"]=label_encoder.fit_transform(sdtes["relationship"])
sdtes["race"]=label_encoder.fit_transform(sdtes["race"])
sdtes["sex"]=label_encoder.fit_transform(sdtes["sex"])

sdtes["native"]=label_encoder.fit_transform(sdtes["native"])

sdtes["Salary"]=label_encoder.fit_transform(sdtes["Salary"])
sdtes
```

```python
x_test=sdtes.iloc[0:200,0:13]
x_test
```

```python
y_test=sdtes.iloc[0:200,13]
y_test
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[33,50,4,10,0.5],'C':[15,14,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 0.1, gamma = 33)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
clf = SVC(kernel= "rbf") 
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)    
```
